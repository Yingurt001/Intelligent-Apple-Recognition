```python
import cv2

def show_color_info(image_path):
    img = cv2.imread(image_path)  # è¯»å–å›¾åƒ
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # è½¬æ¢ä¸º HSV
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)  # è½¬æ¢ä¸º LAB

    # é¼ æ ‡ç‚¹å‡»äº‹ä»¶å¤„ç†
    def on_mouse(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:  # å·¦é”®ç‚¹å‡»
            bgr = img[y, x]
            h, s, v = hsv[y, x]
            l, a, b = lab[y, x]

            print(f"ğŸ–±ï¸ ç‚¹å‡»ä½ç½®: ({x}, {y})")
            print(f"ğŸ¨ BGR: {bgr}")
            print(f"ğŸŒˆ HSV: (H={h}, S={s}, V={v})")
            print(f"ğŸ“˜ LAB: (L={l}, A={a}, B={b})")
            print("-" * 40)

    # æ˜¾ç¤ºçª—å£å¹¶ç»‘å®šç‚¹å‡»å›è°ƒ
    cv2.imshow("ğŸ–¼ï¸ Click to view color", img)
    cv2.setMouseCallback("ğŸ–¼ï¸ Click to view color", on_mouse)

    print("ğŸ’¡ è¯·ç‚¹å‡»å›¾åƒä¸­çš„çº¢è‹¹æœåŒºåŸŸï¼ŒæŸ¥çœ‹é¢œè‰²å€¼ï¼ˆæ§åˆ¶å°è¾“å‡ºï¼‰")
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# ç¤ºä¾‹è°ƒç”¨ï¼ˆä½ å¯ä»¥æ›¿æ¢æˆè‡ªå·±çš„å›¾ç‰‡è·¯å¾„ï¼‰
show_color_info("Attachment_2/1_processed.jpg")
```
